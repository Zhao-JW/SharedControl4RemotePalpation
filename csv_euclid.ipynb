```python
def get_data2():
    # get the distance to ground truth

    dist0_pos = []
    dist1_pos = []
    dist0_orientation = []
    dist1_orientation = []
    angle0 = []
    angle1 = []

    for i, name in enumerate(names[:]):
        # continue
        participant = DataReader(name, dates[i])
        data = participant.readData(invalids[i])
        

        dist0= participant.euclidean_dist(0, isPos=True, forPlot=False)
        dist0_pos.append(np.array(dist0).flatten())
        dist0, angle = participant.euclidean_dist(0, isPos=False, forPlot=False)
        dist0_orientation.append(dist0)
        angle0.append(angle)

        dist1 = participant.euclidean_dist(1, isPos=True, forPlot=False)
        dist1_pos.append(np.array(dist1).flatten())
        dist1, angle = participant.euclidean_dist(1, isPos=False, forPlot=False)
        dist1_orientation.append(dist1)
        angle1.append(angle)
    return dist0_pos, dist1_pos, dist0_orientation, dist1_orientation, angle0, angle1

dist0_pos, dist1_pos, dist0_orientation, dist1_orientation, angle0, angle1 = get_data2()
```

```python
def plot_compare(dist0_pos, dist1_pos):

    # take average like this because not matrix shape
    dist0_pos_ave = []
    dist1_pos_ave = []
    dist0_pos_std = []
    dist1_pos_std = []
    for row in dist0_pos:
        dist0_pos_ave.append(np.average(row)*100)
        dist0_pos_std.append(np.std(row)*100)
    for row in dist1_pos:
        dist1_pos_ave.append(np.average(row)*100)
        dist1_pos_std.append(np.std(row)*100)
    dist_ave = np.vstack((dist0_pos_ave, dist1_pos_ave))

    fig,(ax1, ax2) = plt.subplots(1, 2, figsize=(12,8))
    # plt.title('Position accuracy')
    # position scatter plot
    for i in range(dist_ave.shape[1]):
        ax1.plot([0,1], dist_ave[:2,i])
        ax1.scatter([0,1], dist_ave[:2,i])

    ax1.set_xticks([-0.3,0,1,1.3], ["","Full control", "Shared control",""],fontsize = 18.0)
    ax1.tick_params(labelsize=18)
    # ax1.set_yticks([-0.2,0,0.5,1])
    ax1.set_ylabel('Average distance to ground truth/cm',fontsize = 18.0, labelpad=5)
    ax1.set_title('Position accuracy',fontsize = 18.0)

    diff = dist_ave[1,:]-dist_ave[0,:]
    print(diff)
    ax2.bar(0, np.mean(diff), yerr = np.std(diff), align='center', alpha=0.5, ecolor='black', capsize=10)
    ax2.scatter([0]*10, diff, alpha=0.3)
    ax2.set_xlim([-1,1])
    ax2.set_ylim([-1.3,0.1])
    ax2.set_xticks([0], ["Decrease of distance to ground truth"],fontsize = 18.0)
    ax2.tick_params(labelsize=18)
    ax2.yaxis.tick_right()
    ax2.set_title('P = 6.52e-12',fontsize = 18.0)
    # ax.yaxis.grid(True)

    plt.show()
plot_compare(dist0_pos, dist1_pos)
```
